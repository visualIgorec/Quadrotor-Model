```python
# Разделеные данных на тренировочные и тестовые. Первые 70% строк — тренировочные, остальные — тестовые.
train = data_features.head(int(len(data_features)*0.7))
train
```

```python
test = data_features.tail(int(len(data_features)*0.3))
test
```

```python
train_labels = train['Outcome']
test_labels = test['Outcome']

del train['Outcome']
del test['Outcome']
test
```

```python
# Нормализация данных
from sklearn import preprocessing
normalization = preprocessing.StandardScaler()
normalization.fit(train)
train = normalization.transform(train)
train = pd.DataFrame(data = train)  # Перевод из numpy ---> pandas
normalization.fit(test)
test = normalization.transform(test)
test = pd.DataFrame(data = test)  # Перевод из numpy ---> pandas
```

```python
clf = svm.SVC(gamma='auto', kernel='rbf')
clf.fit(train, train_labels)
```

```python
# Доля верно классифицированных данных на обучающей выборке
from sklearn.metrics import classification_report
predictions_train = clf.predict(train)    
print(classification_report(train_labels, predictions_train))
```

```python
from sklearn.metrics import f1_score
f1_score(train_labels, predictions_train)
```

```python
# Доля верно классифицированных данных на тестовой выборке
predictions_test = clf.predict(test)    
print(classification_report(test_labels, predictions_test))
```

```python
from sklearn.metrics import f1_score
predictions_test = clf.predict(test)
f1_score(test_labels, predictions_test)
```
